```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

train_met = pd.read_csv('training_metrics.csv')
eval_met = pd.read_csv('evaluation_metrics.csv')

### Model Loss
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(train_met['val_loss'], label='val_loss', color='c')
ax.plot(train_met['loss'], label='loss', color='y')
plt.xticks(np.linspace(0,19,20,dtype=int))
fig.legend(loc="upper left")

ax.set_xlabel('epoch')
ax.set_ylabel('loss')

ax.set_title('Model Loss')

plt.savefig('training_met_plotting.png')
plt.show()

### Evaluation
# fig = plt.figure()
# ax = fig.add_subplot(111)

plt.plot(eval_met['mAP(IoU)'], label='mAP', color='b')

plt.legend(loc="upper left")
plt.xticks(np.linspace(0,19,20,dtype=int))
plt.xlabel('epoch')
plt.ylabel('mAP(IoU=0.5)')

plt.title('Model Evaluation')

plt.savefig('eval_met_plotting.png')
plt.show()
```
